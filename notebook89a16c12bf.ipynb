```python
price_zero = df[df['UnitPrice'] == 0]

price_zero[['StockCode', 'Description']].drop_duplicates()

price_negative = df[df['UnitPrice'] < 0]

price_negative[['StockCode', 'Description']].drop_duplicates()

```

```python
cancelaciones = df[df['InvoiceNo'].astype(str).str.startswith('C')]

productos_cancelados = set(cancelaciones['StockCode'])
productos_price_negative = set(price_negative['StockCode'])

anomalias_price = productos_price_negative - productos_cancelados

```

```python
# =========================
# 1. Eliminar duplicados exactos
# =========================
df_clean = df.drop_duplicates().copy()

# =========================
# 2. Separar cancelaciones vÃ¡lidas
# =========================
df_cancel = df_clean[
    df_clean['InvoiceNo'].astype(str).str.startswith('C')
]

# =========================
# 3. Negativos anÃ³malos
# Quantity < 0 SIN cancelaciÃ³n
# =========================
df_negativos_anomalos = df_clean[
    (df_clean['Quantity'] < 0) &
    (~df_clean['InvoiceNo'].astype(str).str.startswith('C'))
]

# =========================
# 4. Precios negativos anÃ³malos
# UnitPrice < 0 SIN cancelaciÃ³n
# =========================
df_price_anomalos = df_clean[
    (df_clean['UnitPrice'] < 0) &
    (~df_clean['InvoiceNo'].astype(str).str.startswith('C'))
]

# =========================
# 5. Eliminar anomalÃ­as
# =========================
indices_remove = set(df_negativos_anomalos.index).union(
    set(df_price_anomalos.index)
)

df_final = df_clean.drop(index=indices_remove).copy()

# =========================
# 6. Etiquetar precios cero
# =========================
df_final['price_zero_flag'] = df_final['UnitPrice'] == 0

# =========================
# 7. Revenue
# =========================
df_final['Revenue'] = (
    df_final['Quantity'] * df_final['UnitPrice']
)

# =========================
# Dataset final
# =========================
df_final
```
